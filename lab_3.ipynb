### Data Preparation

Summary of our dataset indicates the presence of NA values across all columns. Furthermore, it's important to pay special attention to 'prbconv' variable for the following reason. While other variables have 6 NA values, 'prbconv' variable has 5 null values and 1 tilde (special character). A simple na.omit() function will not work for our dataset.

```r
# Simple tail() check on the 'prbconv' variable allowed us to identify the suspect events
print("Closer examination of last 10 values for the 'prbconv' variable")
tail(A$prbconv, 10)
```

Above tail() function shows missing values before the tilde value towards the end of the dataset. Let's take a step back and examine the bottom portion of our dataset.

```r
# Looking at the bottom 10 rows of our dataset
tail(A, 10)
```

Our dataset not only has NA values, but also has incorrect column class for prbconv. In order to address these issues, we will have to transform our dataset.

```r
# Creating a new dataframe without the suspect rows and a new numeric class column for prbconv
B = A[0:91, ]
B$conv_prbconv <- as.numeric(as.character(B$prbconv))
tail(B)
```

NA values are gone and a new column for 'prbconv' has been created at the end of the dataframe as 'prbconv2'. However, there is a duplicate entry (county: 193) in our dataframe. Initially, we wanted to give benefit of the doubt that this could be a valid entry. Unfortunately, closer examination has led us to believe that duplicate entry was the case here. Therefore, we will remove these rows from our new dataframe C. 

```r
# Creating new dataframe C with unique values only
C <- unique(B)
print("Number of samples in dataframe C:")
nrow(C)
```

### Data Analysis

Now that we have a clean dataframe, let's examine our variables with a scatter plot matrix.

```r
# scatterplot matrix to measure colinearity of our variables of interest
scatterplotMatrix(C[ , c("crmrte", "density","conv_prbconv", "prbarr")])
```

Overall, crime rate seems to have a positive relationship with 'density' variable, while having negative relationship with the 'conv_prbconv' variable and the 'prbarr' variable. With this in mind, let's take a closer look at our key variables.

```r
# Examining crmrte
print("Summary of Crime Rate")
summary(C$crmrte)
hist(C$crmrte, col=rgb(1,0,0,0.5), main="Histogram of Crime Rate", xlab="Rate")
box()
```

Again, positive skew with potential outlier values that exceed the density rate of 8. We'll note these observations and move on to the next variables 'conv_prbconv' and 'prbarr'.

It's important to note that 'conv_prbconv' variable and 'prbarr' variable had max values exceeding 1, despite being called 'probability'. Upon further inspection of the data origins, we will conclude that these variables are ratios, not probability. Therefore, we will keep them as they are.

Based on the observations of the histograms above, we can conclude that we are working with non-normally distributed data. Let's go ahead and fit a linear model, predicting crime rate as a function of density.

## Model 1

```r
# Creating the first model
model_zero <- lm(crmrte~density, data=C)
```

```r
# Plotting the model
autoplot(model_zero)
```

```r
# Taking a look at the summary of the model
summary(model_zero)
```

### Variable Transformation with a log()

According to the normal QQ plot, model_zero did not have the best fit. Considering that the 'crmrte' variable was a percentile value, we will go ahead and take a log of the variable for another look at the quality of our model.

```r
# First model with log() on crmrte
model_first <- lm(log(crmrte)~density, data=C)
```

```r
#residual plots for the model with log on crmrte
autoplot(model_first)
```

According to our residuals vs leverage chart for both models, there weren't any significant outliers. However, when we compare the Normal Q-Q plot and Residual vs Fitted between the model with log on crmrte and the model without log on crmrte, it is evident that the model with log on crmrte had better fit.

```r
# Comparing the non-log version and the log version of the first model
stargazer( model_zero, model_first, type = "text", 
           report = "vc*p", # Don't report errors, since we haven't covered them
           title = "Linear Models predicting crime rate",
           keep.stat = c("adj.rsq", "n", "rsq", "f", "ser"),
           omit.table.layout = "n",
           star.cutoffs = c(0.05, 0.01, 0.001)) # Omit more output related to errors
```

The stargazer table above shows that F statistic value was significant for both models. Despite having a lower adjusted R2 value, we will be applying log on crmrte for our first model due to better fit as evidenced in the charts above.

With density as the only factor to our first model, our adjusted R2 was medium with a value of .394 but our F statistic was relatively high with a significant value of 58.843. This is evident from the Residuals vs Fitted chart above that seems to be crowded towards the fitted value of -3.5.

It is also important to note that density has a positive correlation with expectation of crime rate to increase by 22.8 percent, if the density increases by 1 percent.
